```python
import glob

# get data file names
path =r'C:\Users\hassaan\Downloads\DataAnalysis-icodeguru-Batch2-master (1)\DataAnalysis-icodeguru-Batch2-master\Assignments\csvFiles'
filenames = glob.glob(path + "/*.csv")

dfs = []
for filename in filenames:
    dfs.append(pd.read_csv(filename))

# Concatenate all data into one DataFrame
full_data = pd.concat(dfs, ignore_index=True)
```

## Organizing Data for Plotting

```python
years=full_data['year'].unique()
male_births=total_births_by_sex_and_year['M']
female_births=total_births_by_sex_and_year['F']
```

```python
fig = plt.figure()  # create a plot figure
# plt.rcParams["figure.figsize"]=(14,5)
plt.subplot(2, 2, 1) # (rows, columns, panel number)
plt.plot(years,male_births,'x--r',label="Male Births")
plt.subplot(2, 2, 2)
plt.plot(years,female_births,'o--g',label="Female Births")
plt.subplot(2, 2, 3)
plt.plot(years,male_births,'x--r',label="Male Births")
plt.plot(years,female_births,'o--g',label="Female Births")
plt.legend()
```

```python
plt.rcParams["figure.figsize"]=(14,5)
plt.plot(years,male_births,'x--r',label="Male Births")
plt.plot(years,female_births,'o--g',label="Female Births")
plt.title("Comparision of Male and Feamale births from 1991 to 2010")
plt.xlabel("Year of birth")
plt.ylabel("Total Births")
plt.xticks(years)
plt.legend(fontsize=12, loc="upper center")
```
